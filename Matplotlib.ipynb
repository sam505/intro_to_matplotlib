```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prices = [12, 54, 65, 123, 64, 457, 98]
cash = [76, 34, 66, 98, 12, 65, 43]
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
```

```python
plt.plot(year, prices, 'k',
        year, cash, '--')
plt.scatter(year, cash)

plt.xlabel('Year')
plt.ylabel('Prices')
plt.axis([2010, 2020, -50, 500]) # Define the limits of the axis
plt.show()
```
